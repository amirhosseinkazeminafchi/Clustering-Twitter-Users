```python
#function to update pickle object
import pickle
def pickle_obj(obj, key, path): 
    load_file = open(path, 'rb')
    url_dict = pickle.load(load_file)

    url_dict[key].append(obj)

    update_file = open(path, "wb")
    pickle.dump(url_dict, update_file)
    update_file.close()
```
