```python
file_path = 'data/all_user_info.csv'
users = pd.DataFrame()

import json

ids = []
screen_names = []
locations = []
tweet_ids = []

for i in range(r2.shape[0]): 
    print(i)
    j = json.loads(r2.r_objects[i])

    for user in j:  
        ids.append(user['id'])
        screen_names.append(user['screen_name'])
        locations.append(user['location'])
        tweet_ids.append(r2.tweet_id[i])


users['user_id'] = ids
users['screen_name'] = screen_names
users['location'] = locations
users['tweet_id'] = tweet_ids

```

```python
print(users.shape)
users.head(2)
```

```python
r2.columns
```

```python
u_dict = {}
u_d_screen_names = {}
u_d_location = {}
for i in range(users.shape[0]):
    if not users.user_id[i] in u_dict: 
        u_dict[users.user_id[i]] = [users.tweet_id[i]]
        u_d_screen_names[users.user_id[i]] = [users.screen_name[i]]
        u_d_location[users.user_id[i]] = [users.location[i]]
    else: 
        u_dict[users.user_id[i]].append(users.tweet_id[i])
```

```python
len(u_dict), len(u_d_screen_names), len(u_d_location)
```

```python
u_dict.keys()==u_d_screen_names.keys()==u_d_location.keys()
```

```python
unique_users = pd.DataFrame()
unique_users['user_id'] = list(u_dict.keys())
unique_users['screen_name'] = list(u_d_screen_names.values())
unique_users['location'] = list(u_d_location.values())
unique_users['retweet_list'] = list(u_dict.values())
```

```python
unique_users['retweet_count'] = [len(x) for x in unique_users['retweet_list']]
unique_users.head()
```

```python
unique_users.to_csv('data/unique_users.csv')
```
