```python
def get_article(url):
    a = Article(url, language='en')
    a.download()
    a.parse()
    return a
```

```python

import time

file_path = 'data/art_info.csv'
df1 = pd.DataFrame(columns = ['tweet_id','text', 'top_image', 'title', 'authors', 'summary', 'keywords'])
df1.to_csv(file_path)

for i  in range(ny.shape[0]):
    t1 = time.time()
    print(i)
    url = ny.final_url[i]
    time.sleep(0.5)
    mini_df = pd.DataFrame()
    mini_df['tweet_id'] = [ny['tweet_id'][i]]
    
    try: 
        article = get_article(url)
        mini_df['text'] = [article.text]
        mini_df['top_image'] = [article.top_image]
        mini_df['title'] = [article.title]
        mini_df['authors'] = [article.authors]
        article.nlp()
        mini_df['summary'] = [article.summary]
        mini_df["keywords"] = [article.keywords]

    except Exception as e: 
        print('exception at index:', i, e)
        mini_df['text'] = np.nan
        mini_df['top_image'] = np.nan
        mini_df['title'] = np.nan
        mini_df['authors'] = np.nan
        article.nlp()
        mini_df['summary'] = np.nan
        mini_df["keywords"] = np.nan
    
    mini_df.to_csv(file_path, mode='a', header=False)
    t2 = time.time()
    print(t2-t1, 'seconds')
```

```python
art_info = pd.read_csv(file_path)
```
